```python
from autoscraper import AutoScraper

url = 'https://github.com/krishnaik06?tab=repositories'

#put unique object information from a page that need to be scraped

wanted_list = ['103', 'Malaria-Detection']

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
from autoscraper import AutoScraper
import pandas as pd

url = 'https://github.com/avinassh?tab=repositories'

#put unique object information from a page that need to be scraped

wanted_list = ['488', 'grpc-errors']

scraper = AutoScraper()
result = scraper.build(url, wanted_list)

df = pd.DataFrame()
af = pd.DataFrame()
scraper.build(url, [wanted_list[0]])
df["Stars"] = scraper.get_result_similar(url, unique=False, keep_order=True)
scraper.build(url, [wanted_list[1]])
af["Title"] = scraper.get_result_similar(url, unique=False, keep_order=True)
```

```python
dframe = pd.concat([df, af], axis=1)
```
